```python
from bertopic import BERTopic
import ast
import pandas as pd
```

```python
# Read input file and create table
records = []
with open("data/data_mining_publications.txt", "r") as f:
    for line in f:
        line = line.strip()
        if line:
            records.append(ast.literal_eval(line))


table = pd.DataFrame(records)

# Get all titles and dates
titles = table['title'].to_list()
timestamps = table['year'].to_list()
```

```python
from bertopic.representation import KeyBERTInspired

# key_bertopic_model.topics_over_time() increases the coherence and 
# reduces stopwords from the resulting topic representations
representation_model = KeyBERTInspired()
topic_model = BERTopic(representation_model=representation_model)
topic_model.fit_transform(titles)
```

```python
topics_over_time = topic_model.topics_over_time(titles, timestamps)
```

```python
topic_model.visualize_topics_over_time(topics_over_time, top_n_topics=10)
```
